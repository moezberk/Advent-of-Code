```python
with open ("door1.txt", "r") as f_in: #öffnet strom
    lines = f_in.readlines() #alle zeilen werden ausgelesen in der variable lines

alle_paare=[]
for line in lines: #iteriert über alle zeilen
    zahlen_paar=""
    for character in line:
        if character.isdigit():
            zahlen_paar+=character
            break
    for character in line[::-1]:
        if character.isdigit(): #drittes Argument gibt Schrittlänge an: Start, Ende, Schrittlänge
            zahlen_paar+=character
            break
    alle_paare.append(zahlen_paar)
print(alle_paare)
endergebnis= 0
for paar in alle_paare:
    endergebnis +=int(paar)
print(endergebnis)
```
